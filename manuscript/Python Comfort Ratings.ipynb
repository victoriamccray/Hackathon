## Slope Chart

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# Data
ratings = [1, 2, 3, 4, 5]
before = [49.50, 18.40, 21.70, 6.10, 4.20]  # Interest Form Applicants (%)
after = [0.00, 0.00, 14.30, 50.00, 35.70]  # Hackathon Participants (%)

# Convert to DataFrame for easier handling
df = pd.DataFrame({"Rating": ratings, "Before": before, "After": after})
```

```python
# Plot
plt.figure(figsize=(8, 5))
for i in range(len(ratings)):
    plt.plot([0, 1], [df["Before"][i], df["After"][i]], marker="o", color="black", lw=2)
    plt.text(-0.1, df["Before"][i], f"{df['Before'][i]:.1f}%", ha="right", va="center", fontsize=12)
    plt.text(1.1, df["After"][i], f"{df['After'][i]:.1f}%", ha="left", va="center", fontsize=12)

# Aesthetics
plt.xticks([0, 1], ["Before Hackathon", "After Hackathon"], fontsize=12)
plt.ylabel("Percentage of Respondents", fontsize=12)
plt.title("Shift in Python Comfort Levels", fontsize=14, fontweight="bold")
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Show Plot
plt.show()
```

## Diverging Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
ratings = [1, 2, 3, 4, 5]
before = [-49.50, -18.40, -21.70, -6.10, -4.20]  # Negative for left alignment
after = [0.00, 0.00, 14.30, 50.00, 35.70]  # Positive for right alignment

# Plot
fig, ax = plt.subplots(figsize=(8, 5))
ax.barh(ratings, before, color="#1f77b4", label="Before Hackathon")
ax.barh(ratings, after, color="#ff7f0e", label="After Hackathon")

# Labels and Style
ax.set_yticks(ratings)
ax.set_yticklabels([f"Comfort Level {r}" for r in ratings])
ax.set_xlabel("Percentage of Respondents")
ax.set_title("Python Confidence Shift: Before vs. After Hackathon", fontsize=14, fontweight="bold")
ax.axvline(0, color="black", lw=1)  # Center line

plt.legend()
plt.show()
```

## Stacked Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ["Before", "After"]
data = [
    [49.50, 18.40, 21.70, 6.10, 4.20],  # Before
    [0.00, 0.00, 14.30, 50.00, 35.70]  # After
]
colors = ["#d73027", "#fc8d59", "#fee08b", "#91cf60", "#1a9850"]

# Plot
fig, ax = plt.subplots(figsize=(8, 5))
bottom = np.zeros(2)

for i, (color, level) in enumerate(zip(colors, [1, 2, 3, 4, 5])):
    ax.bar(labels, [data[0][i], data[1][i]], color=color, label=f"Level {level}", bottom=bottom)
    bottom += np.array([data[0][i], data[1][i]])

# Labels
ax.set_title("Python Comfort Levels Before & After Hackathon", fontsize=14, fontweight="bold")
ax.set_ylabel("Percentage of Respondents")
plt.legend(title="Comfort Level")

plt.show()
```

## Heatmap

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data
df = pd.DataFrame({
    "Before": [49.50, 18.40, 21.70, 6.10, 4.20],
    "After": [0.00, 0.00, 14.30, 50.00, 35.70]
}, index=[1, 2, 3, 4, 5])

# Heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(df, annot=True, cmap="coolwarm", fmt=".1f", linewidths=0.5)

# Labels
plt.xlabel("Survey Stage")
plt.ylabel("Python Comfort Level")
plt.title("Python Confidence Progression (Before vs. After Hackathon)", fontsize=14, fontweight="bold")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
ratings = [1, 2, 3, 4, 5]
before = [49.50, 18.40, 21.70, 6.10, 4.20]  # Interest Form Applicants (%)
after = [0.00, 0.00, 14.30, 50.00, 35.70]  # Hackathon Participants (%)

# Bar width and positions
bar_width = 0.4
x = np.arange(len(ratings))

# Plot
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x - bar_width/2, before, bar_width, label="Before Hackathon", color="#1f77b4")
ax.bar(x + bar_width/2, after, bar_width, label="After Hackathon", color="#ff7f0e")

# Labels & Style
ax.set_xticks(x)
ax.set_xticklabels(ratings)
ax.set_xlabel("Python Comfort Level", fontsize=12)
ax.set_ylabel("Percentage of Respondents", fontsize=12)
ax.set_title("Comparison of Python Confidence Before & After Hackathon", fontsize=14, fontweight="bold")

# Add percentage labels
for i in range(len(ratings)):
    ax.text(x[i] - bar_width/2, before[i] + 1, f"{before[i]:.1f}%", ha="center", fontsize=10, color="black")
    ax.text(x[i] + bar_width/2, after[i] + 1, f"{after[i]:.1f}%", ha="center", fontsize=10, color="black")

# Add legend and place it outside the plot for better visibility
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))

# Show grid for readability
ax.yaxis.grid(True, linestyle="--", alpha=0.5)

# Adjust layout to fit everything
plt.tight_layout()

plt.savefig("python_comfort_levels.png", dpi=300, bbox_inches="tight")
#plt.savefig("python_comfort_levels.png")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patheffects as pe

# Data
ratings = [1, 2, 3, 4, 5]
before = [49.50, 18.40, 21.70, 6.10, 4.20]
after = [0.00, 0.00, 14.30, 50.00, 35.70]

# Bar width and positions
bar_width = 0.35
x = np.arange(len(ratings))

# Modern color palette (colorblind accessible)
before_color = "#5DA5DA"  # Blue
after_color = "#FAA43A"   # Orange

# Plot
fig, ax = plt.subplots(figsize=(6, 6))  # square aspect ratio
bars1 = ax.bar(x - bar_width/2, before, bar_width, label="Before", color=before_color, edgecolor="none")
bars2 = ax.bar(x + bar_width/2, after, bar_width, label="After", color=after_color, edgecolor="none")

# Text and Title
ax.set_title("Python Confidence\nBefore vs. After Hackathon", fontsize=16, fontweight="bold", pad=20)
ax.set_xlabel("Comfort Level (1–5)", fontsize=12)
ax.set_ylabel("Percent of Respondents", fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(ratings, fontsize=11)
ax.tick_params(axis='y', labelsize=11)

# Add percentage labels with shadow
for bar in bars1 + bars2:
    height = bar.get_height()
    if height > 0:
        ax.text(
            bar.get_x() + bar.get_width()/2,
            height + 1,
            f"{height:.1f}%",
            ha='center',
            va='bottom',
            fontsize=10,
            color='black',
            path_effects=[pe.withStroke(linewidth=2, foreground="white")]
        )

# Grid and Legend
ax.yaxis.grid(True, linestyle="--", alpha=0.4)
ax.set_axisbelow(True)
ax.legend(loc="upper center", bbox_to_anchor=(0.5, -0.12), ncol=2, frameon=False, fontsize=11)

# Minimal layout adjustments
plt.tight_layout(pad=2)

# Save
plt.savefig("python_comfort_social.png", dpi=300, bbox_inches="tight", facecolor="white")
plt.show()
```
